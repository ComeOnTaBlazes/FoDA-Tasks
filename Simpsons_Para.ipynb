```python

n_samples = 100
np.random.seed(42)
ages18 = rng.integers(18, 33, n_samples)
cigs18 = rng.integers(1, 18, n_samples) - rng.normal(5, 4 ,n_samples)
a = 12
b = 10
```

```python
y  = (a * ages18 + b * (cigs18 + rng.normal(2, 0.5, len(cigs18))))/1000
```

```python
y
```

```python
plt.plot(ages18, y, '.', label = '18 vs prob of disease')
```

```python
ages33 = rng.integers(33, 48, n_samples)
cigs33 = rng.integers(4, 20, n_samples) + rng.normal(1, 3 ,n_samples)
a = 9
b = 7
y1  = (a * (ages33*0.75) + b*(cigs33 + rng.normal(2, 0.5, len(cigs33))))/1000
```

```python
y1
```

```python
plt.plot(ages18, y, '.', label = '18 vs prob of disease')
plt.plot(ages33, y1, '.', label = '28 vs prob of disease')
plt.legend()
```

```python
ages48 = rng.integers(48, 63, n_samples)
cigs48 = rng.integers(1, 15, n_samples) + rng.normal(1, 3 ,n_samples)
a = 7
b = 5
y2  = (a * (ages48*0.50) + b*(cigs48 + rng.normal(2, 0.5, len(cigs48))))/1000
```

```python
plt.plot(ages18, y, '.', label = '18 vs prob of disease')
plt.plot(ages33, y1, '.', label = '28 vs prob of disease')
plt.plot(ages48, y2, '.', label = '28 vs prob of disease')
plt.legend()
```

```python
ages63 = rng.integers(63, 78, n_samples)
cigs63 = rng.integers(1, 30, n_samples) + rng.normal(7, 3 ,n_samples)
a = 6
b = 5
y3  = (a * (ages63*0.25) + b*(cigs63 + rng.normal(3, 0.75, len(cigs63))))/1000
```

```python
plt.plot(ages18, y, '.', label = '18 vs prob of disease')
plt.plot(ages33, y1, '.', label = '33 vs prob of disease')
plt.plot(ages48, y2, '.', label = '48 vs prob of disease')
plt.plot(ages63, y3, '.', label = '63 vs prob of disease')

plt.legend()
```
