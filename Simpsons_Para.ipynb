```python

np.random.seed(42)
def smoker_data(n_samples, age_low, age_hig, cig_low, cig_high):
    n_samples = n_samples
    age = rng.integers(age_low, age_hig, n_samples)
    cigs = rng.integers(cig_low, cig_high, n_samples) - rng.normal(5, 4 ,n_samples)
    return(pd.DataFrame(age ,cigs))
```

```python
n_samples = 100
```

```python
def x_axis(start, end):
    ans = np.linspace(start,end, num=n_samples)
    return ans
```

```python
def y_axis(a, b ,x):
    y= a*x + (b - + rng.normal(2, 2, n_samples))
    return y
```

```python
x1 = x_axis(1,4)
x2 = x_axis(4,8)
x3 = x_axis(8,12)
x4 = x_axis(12,16)
```

```python
y1 = y_axis(5,7, x1)
y2 = y_axis(3,0, x2)
y3 = y_axis(2, -2, x3)
y4 = y_axis(1, -4, x4)
```

```python
plt.plot(x1, y1, '.')
plt.plot(x2, y2, '.')
plt.plot(x3, y3, '.')
plt.plot(x4, y4, '.')
```

```python
df = pd.DataFrame({'x_axis': x1, 'y_axis': y1, 'group': 'A'})
df2 = pd.DataFrame({'x_axis': x2, 'y_axis': y2, 'group': 'B'})
df3 = pd.DataFrame({'x_axis': x3, 'y_axis': y3, 'group': 'C'})
df4 = pd.DataFrame({'x_axis': x4, 'y_axis': y4, 'group': 'D'})
allsets= [df, df2, df3,df4]
```

```python
combined= pd.concat(allsets, keys = [ 'a', 'b', 'c', 'd'])
```

```python
sns.relplot(data= combined, x = 'x_axis', y = 'y_axis', hue = 'group')
```
