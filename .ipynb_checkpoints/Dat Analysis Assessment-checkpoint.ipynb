Ok, the desired output has been achieved, this should be pretty simple to put into a function now

```python
def counts(x):
    keys = []
    countval = []
    for val in x:
        if val not in keys:
            keys.append(val)
    for n in keys:
        ans = x.count(n)
        countval.append(ans)
    dictout = dict(zip(keys, countval))
    return dictout
```

```python
counts(input)
```

```python
counts(['A', 'A', 'B', 'C', 'A'])    
```
